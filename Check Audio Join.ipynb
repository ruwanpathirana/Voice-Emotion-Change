```python
from pydub import AudioSegment
import os
from IPython.display import Audio 

audio_folder = "Audio"

# Load the two audio files

audio_file1 = AudioSegment.from_wav(os.path.join(audio_folder, "file1.wav"))
audio_file2 = AudioSegment.from_wav(os.path.join(audio_folder, "file2.wav"))
```

```python
# Join the audio files

joined_audio = audio_file1 + audio_file2
```

```python
# Export the joined audio to a new file

output_file_path = os.path.join(audio_folder, "joined_audio.wav")
joined_audio.export(output_file_path, format="wav")
```

```python
Audio(output_file_path, autoplay=True)
```
